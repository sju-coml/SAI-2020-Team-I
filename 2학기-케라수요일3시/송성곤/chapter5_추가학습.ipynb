dst	=	cv.Sobel(	src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]]	)
- dx + dy ==1
- src : input_img
- dst : output img (same size, same channel)
- ddepth :output depth
- dx : order of the deriverative x 
- dy : order of the deriverative y
- ksize : size of extended Sobel kernel; it must be 1,3,5, or 7.
- scale : optional scale factor for the computed derivative values
- delta : optional delta value that is added to the results prior to string them in dst

```python
def sobel(file):
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=-1)
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=-1)
    plt.figure(1,figsize=(17,10))
    plt.subplot(2,2,1),plt.imshow(img,cmap='gray')
    plt.title('Original'), plt.xticks([]),plt.yticks([])

    plt.subplot(2,2,3),plt.imshow(sobelx,cmap='gray')
    plt.title('Sobel X'), plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,4),plt.imshow(sobely,cmap='gray')
    plt.title('Sobel Y'), plt.xticks([]),plt.yticks([])
    plt.show
sobel('road1.jpg')
```

```python
def sobelxy(file):
    img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    sobelx = cv2.Sobel(img, cv2.CV_64F,1,0,ksize=-1)
    sobely = cv2.Sobel(img, cv2.CV_64F,0,1,ksize=-1)
    return sobelx, sobely
```

```python
img = cv2.imread('road2.jpg',cv2.IMREAD_GRAYSCALE)
sobelx = cv2.Sobel(img, cv2.CV_64F,1,0,ksize=-1)
sobely = cv2.Sobel(img, cv2.CV_64F,0,1,ksize=-1)
```

```python
sobel('road2.jpg')
```

```python
sobel_G = np.sqrt(sobelx*sobelx + sobely*sobely)
theta = np.arctan2(sobely,sobelx)
```

```python
TRHESHGOLD_ORIENTATION = -1
mask1 = (theta > -(np.pi/2))
mask2 = (theta <0)

mask_12 = np.logical_and(mask1,mask2)
new_theta = np.multiply(mask_12,theta)
```
